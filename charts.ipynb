```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotCounts(df, axx, title, hide_y):
    ax = df.plot(ax=axx, kind='barh', title=title)
    ax.bar_label(ax.containers[0], fontsize=16)
    ax.bar_label(ax.containers[1], fontsize=16)
    ax.axes.xaxis.set_visible(False)
    if(hide_y):
        ax.axes.yaxis.set_visible(False)
    ax.get_legend().remove()
    for spine in ['top', 'right', 'bottom']:
       ax.spines[spine].set_visible(False)

idx = ['Major','Minor','Rejected']
gl = 'GitLab'
ots = 'over-the-shoulder'
e1 = pd.DataFrame([{gl:3,ots:7}, {gl:266,ots:142}, {gl:98,ots:2}], index=idx)
e2 = pd.DataFrame([{gl:1,ots:1}, {gl:48,ots:20}, {gl:12,ots:1}], index=idx)
e3 = pd.DataFrame([{gl:6, ots:11}, {gl:238,ots:142}, {gl:71,ots:5}], index=idx)

fig, axes = plt.subplots(1, 3, figsize=(22, 5))
fig.suptitle('Fig. 1 Comments count', fontsize=16, y=0.02)

plotCounts(e1, axes[0], 'Experiment 1', False)
plotCounts(e2, axes[1], 'Experiment 2', True)
plotCounts(e3, axes[2], 'Experiment 3', True)
fig.legend([gl,ots], loc=(0.86,0.17))
fig.savefig('counts.png', bbox_inches="tight")
```

```python
def plotProportions(df, axx, title, hide_y):
    ax = df.plot.barh(ax=axx, stacked=True, color=['tab:green', 'tab:blue', 'tab:red'], title=title)
    ax.bar_label(ax.containers[0], fontsize=16, fmt='%d%%', padding=-3)
    ax.bar_label(ax.containers[1], fontsize=16, fmt='%d%%', label_type='center')
    ax.bar_label(ax.containers[2], fontsize=16, fmt='%d%%', label_type='center')
    ax.axes.xaxis.set_visible(False)
    if(hide_y):
        ax.axes.yaxis.set_visible(False)
    ax.get_legend().remove()
    for spine in ['top', 'right', 'bottom', 'left']:
       ax.spines[spine].set_visible(False)
    
idx = ['GitLab', 'over-the-shoulder']
maj = 'Major'
mi = 'Minor'
r = 'Rejected'
e1 = pd.DataFrame([{maj:1, mi:72, r:27}, {maj:5, mi:94, r:1}], index=idx)
e2 = pd.DataFrame([{maj:2, mi:79, r:20}, {maj:5, mi:90, r:5}], index=idx)
e3 = pd.DataFrame([{maj:2, mi:76, r:23}, {maj:7, mi:90, r:3}], index=idx)

fig, axes = plt.subplots(1, 3, figsize=(21, 5))
fig.suptitle('Fig. 2 Proportions of different comments types', fontsize=16, y=0.02)
plotProportions(e1,  axes[0], 'Experiment 1', False)
plotProportions(e2,  axes[1], 'Experiment 2', True)
plotProportions(e3,  axes[2], 'Experiment 3', True)

fig.legend([maj,mi,r], loc=(0.89,0.12))
fig.savefig('proportions.png', bbox_inches="tight")
```

```python
def plotKnowledge(df, axx, title, hide_y):
    ax = df.plot(ax=axx, kind='barh', title=title, color=['tab:green'])
    ax.axes.xaxis.set_visible(False)
    if(hide_y):
        ax.axes.yaxis.set_visible(False)
    ax.get_legend().remove()
    for spine in ['top', 'right', 'bottom']:
       ax.spines[spine].set_visible(False)
    
idx = ['It does not support knowledge sharing', 'It supports knowledge sharing to a limited extent', 'It is good at supporting knowledge sharing, but I see areas where it could be improved', 'It is hard to imagine a better way to support knowledge sharing']
gl = 'GitLab'
ots = 'over-the-shoulder'
df1 = pd.DataFrame([{ots:0}, {ots:1}, {ots:10}, {ots:7}], index=idx)
df2 = pd.DataFrame([{gl:1}, {gl:13}, {gl:3}, {gl:1}], index=idx)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Fig. 3 Knowledge transfer effectiveness assessment - all experiments', fontsize=16, y=0.02)
plotKnowledge(df1,  axes[0], ots, False)
plotKnowledge(df2,  axes[1], gl, True)

fig.savefig('knowledge.png', bbox_inches="tight")
```
